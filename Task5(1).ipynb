```
fig, axes = plt.subplots(figsize = (10 , 10))
sns.boxplot(stores_df['Sales'])
```

```
sales_by_region = stores_df.groupby('Region').size()
print(sales_by_region)
```

```


sales_by_region = sales_by_region.plot(title='Sales by Region',xticks=(1,2,3,4))
sales_by_region.set_xlabel('Region')
sales_by_region.set_ylabel('Sales')

```

```
sales_by_category = stores_df.groupby('Category').size()
print(sales_by_category)
```

```
plot_by_category = sales_by_category.plot.bar(title='Sales by Category')
plot_by_category.set_xlabel('Category')
plot_by_category.set_ylabel('Sales')
```

```
sales_segment = stores_df.groupby('Segment').size()
print(sales_segment)
```

```
sales_segment = sales_segment.plot.bar(title='Sales by Segment')
sales_segment.set_xlabel('Segment')
sales_segment.set_ylabel('Sales')
```
